### 4.3.1 微分

```python
def function_1(x):
    w_1 = 0.01 
    w_2 = 0.1
    b = 0
    y = w_1*(x**2) + w_2*(x)
    
    return y
```

```python

x = np.arange(0.0, 20.0, 0.1) #從0-20，以0.1為單位的x陣列
y = function_1(x)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x, y)
plt.show()
```

#### 計算微分

```python
def numerical_diff(f,x):
    h = 1e-5 #0.0001
    return (f(x+h) -f(x-h)) / (2*h)

# numerical_diff(function_1, 5)
numerical_diff(function_1, 10)
```

#### 繪製切線斜率

```python

def tangent_line(f, x):
    d = numerical_diff(f, x)
    print('d : ',d)
    b = f(x) - d*x

    print('b : ',b)
    return lambda t: d*t + b

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")

tf = tangent_line(function_1, 10)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
plt.show()

# [_ for _ in zip(x, y2)] 
```

### 4.3.3 偏微分

```python
# f(x0 , x1) = x0^2 + x1^2

def function_2(x):  # 輸入陣列
    if x.ndim == 1:
        return np.sum(x**2)  # 等同於 x[0]**2 + x[1]**2
    else:
        return np.sum(x**2, axis=1)
```
